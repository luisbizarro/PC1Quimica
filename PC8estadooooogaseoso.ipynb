```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


datos = {'Categoría': ['Hidrógeno', 'Helio', 'Litio', 'Berilio', 'Boro', 'Carbono', 'Nitrógeno', 'Oxi­geno', 'Fluor', 'Neón	'],
        'Valor': [1.0079, 4.0026, 6.941, 9.0122, 10.811, 12.0107, 14.0067, 15.9994, 18.9984, 20.1797]}
df = pd.DataFrame(datos)

# grafico de barras
plt.figure(figsize=(10, 8))
sns.barplot(x='Categoría', y='Valor', data=df)

# Agregar etiquetas y título
plt.xlabel('Elementos')
plt.ylabel('Peso atómico')
plt.title('Diagrama de Barras de los pesos atomicos de algunos elementos de la tabla periódica')

# Mostrar el gráfico
plt.show()
```
